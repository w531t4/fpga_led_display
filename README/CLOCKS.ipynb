```python

GOAL_PANEL_REFRESH_RATE_HZ = 30
GOAL_PANEL_REFRESH_RATE_PERIOD_SECS = 1/GOAL_PANEL_REFRESH_RATE_HZ
GOAL_PANEL_REFRESH_RATE_PERIOD_USECS = GOAL_PANEL_REFRESH_RATE_PERIOD_SECS*1000000

print("GOAL_PANEL_REFRESH_RATE_HZ = %s hz" % GOAL_PANEL_REFRESH_RATE_HZ)
print("GOAL_PANEL_REFRESH_RATE_PERIOD_SECS = {secs:.9f} seconds".format(secs=GOAL_PANEL_REFRESH_RATE_PERIOD_SECS))
print("GOAL_PANEL_REFRESH_RATE_PERIOD_USECS = {usecs:.5f} microseconds".format(usecs=GOAL_PANEL_REFRESH_RATE_PERIOD_USECS))
```

The 32x64 panel is actually 2x 16x64 panels, with a shared row selector bus (`row_adddress_actve[0:3]`), but independent coloring buses (`rgb1`, `rgb2`).

We'll need to iterate through the 16 rows within `GOAL_REFRESH_RATE_PERIOD_SECS`


```python

GOAL_PERROW_REFRESH_RATE_HZ = GOAL_PANEL_REFRESH_RATE_HZ*16
GOAL_PERROW_REFRESH_RATE_PERIOD_SECS = GOAL_PANEL_REFRESH_RATE_PERIOD_SECS / 16
GOAL_PERROW_REFRESH_RATE_PERIOD_USECS = GOAL_PANEL_REFRESH_RATE_PERIOD_USECS / 16

print("GOAL_PERROW_REFRESH_RATE_HZ = %s hz" % GOAL_PERROW_REFRESH_RATE_HZ)
print("GOAL_PERROW_REFRESH_RATE_PERIOD_SECS = {secs:.9f} seconds".format(secs=GOAL_PERROW_REFRESH_RATE_PERIOD_SECS))
print("GOAL_PERROW_REFRESH_RATE_PERIOD_USECS = {usecs:.5f} microseconds".format(usecs=GOAL_PERROW_REFRESH_RATE_PERIOD_USECS))
```

We need to write 64 pixels per row

```python

GOAL_PERPIXEL_REFRESH_RATE_HZ = GOAL_PERROW_REFRESH_RATE_HZ*64
GOAL_PERPIXEL_REFRESH_RATE_PERIOD_SECS = GOAL_PERROW_REFRESH_RATE_PERIOD_SECS / 64
GOAL_PERPIXEL_REFRESH_RATE_PERIOD_USECS = GOAL_PERROW_REFRESH_RATE_PERIOD_USECS / 64

print("GOAL_PERPIXEL_REFRESH_RATE_HZ = %s hz" % GOAL_PERPIXEL_REFRESH_RATE_HZ)
print("GOAL_PERPIXEL_REFRESH_RATE_PERIOD_SECS = {secs:.9f} seconds".format(secs=GOAL_PERPIXEL_REFRESH_RATE_PERIOD_SECS))
print("GOAL_PERPIXEL_REFRESH_RATE_PERIOD_USECS = {usecs:.5f} microseconds".format(usecs=GOAL_PERPIXEL_REFRESH_RATE_PERIOD_USECS))
```
